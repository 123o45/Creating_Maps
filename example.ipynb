```python
import pandas as pd
import json
```

```python
# Open the file with 'utf-8-sig' encoding to handle BOM characters
with open("world.json", "r", encoding="utf-8-sig") as file:
    raw_content = file.read()  # Read the raw content
    #print(f"Raw content: {raw_content}")  # This should no longer include the BOM

try:
    content = json.loads(raw_content)

    # Normalize the JSON structure and create a DataFrame
    df = pd.json_normalize(content['features'])  # Adjust based on the JSON structure
    df.head()
except json.JSONDecodeError as e:
    print(f"JSON Decode Error: {e}")

```
